```python
import numpy as np
import pandas as pd
```

```python
# 1. 데이터 불러오기(read_csv) DataFrame으로 불러옴
data = pd.read_csv('data/BOAZ_info.csv')
type(data)
```

```python
# 2. 데이터 확인( .head(), shape, columns, describe())
print(data.head())
print(data.shape)
data.describe()
```

```python
# 3. ("number")  9기만 다루는 DataFrame 구성(loc) -> 확인(head) -> 인덱스 번호 다시 매겨주기(index)
data2 = data.loc[data['number']==9,]
print(data2.head())
#data2 = data2.reindex(range(data2.shape[0]))
data2.index = np.arange(data2.shape[0])
data2.tail()
```

```python
# 4. ("name") 9기 이름 뽑아서  "BOAZ_9_name"이라는 변수에 저장 
BOAZ_9_name = data2.name
```

```python
# 5. ("laptop") 불필요한 열 제거(drop / del) -> 확인(shape, head)
data2.drop('laptop',axis = 1)
#del data2['laptop']
data2.head()
```

```python
# 6. ("weight")결측치 있나 확인(isnull) -> 결측치 다 60으로 채워버리기(fillna)-> 확인(isnull)
data2.weight.isnull()
data2.weight = data2.weight.fillna(value=60)
data2.isnull()
```

```python
# 7. ("age") 통계치 뽑아보기(min, max, mean)
print(data2.age.min())
print(data2.age.max())
print(data2.age.mean())
```

```python
# 8. ("attendance") 출석율 0.3 이하인 사람 탈동시키기ㅠㅠ(loc) ->  확인(shape) -> 인덱스 번호 다시 붙여주기(index)
data2 = data2[data2.attendance>0.3]
data2.index = np.arange(data2.shape[0])
data2.tail()
```

```python
# 9. ("passion") boolean mask 출석율 0.7이상이면 T, 아니면 F  -> 확인
data2['passion'] = data2['attendance'] >= 0.7 # data2.passion = data2.attendence >= 0.7 안먹힘// 왜??
data2.head()
```

```python
# 10. ("blood_type") 무슨 혈액형이 제일 많나(value_counts)
data2.blood_type.value_counts()
```

```python
# 11. ("alcohol") 나이와 주량 상관계수 구해보기(corr)
data2.alcohol.corr(data2.age)
```

```python
# 12. ("quiz1") 평균(mean)
data2.quiz1.mean()
```

```python
# 13. ("quiz2") 개인별 quiz 1 & 2 더해 "quiz_sum" 시리즈 변수 만들고 -> 시험등수 열 'quiz_ranking'만들기(rank) -> 'quiz_ranking'별로 정렬해보기(sort_values)
quiz_sum = data2.quiz1+data2.quiz2;type(quiz_sum)
data2['quiz_ranking'] = quiz_sum.rank(ascending=False)
data2 = data2.sort_values(by='quiz_ranking',ascending=True)
```

```python
#14. ‘BOAZ_9_Info’ csv로 저장하기(디렉토리 확인!)
data2.to_csv('BOAZ_9_Info.csv',sep=',')
```
