```python
import requests
 
response = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn')
 
html = response.text


```

```python
from bs4 import BeautifulSoup    #BeautifulSoup import
soup = BeautifulSoup(html, 'html.parser') #html.parser를 사용해서 soup에 넣겠다
tag1 = soup.select('div[class=tit3]')

```

```python
for tag in soup.select('div[class=tit3]'):
    print(tag.text.strip())
```

```python
print(soup.find('div', attrs={'class':'tit3'}).text)
```

```python
print(soup.find_all('div',{'class' : 'tit3'}))
```

```python
for tag in soup.find_all('div',{'class' :'tit3'}):
    print (tag.text.strip())
```

```python
response = requests.get('https://music.bugs.co.kr/chart/track/day/total?chardate=20180523')
html = response.text
soup = BeautifulSoup(html, 'html.parser') 

for link in soup.find_all('p',{'class':'title'}):
    print(link.text.strip())
```
