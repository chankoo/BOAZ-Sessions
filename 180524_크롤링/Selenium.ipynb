```python
#pip install selenium
from selenium import webdriver
from bs4 import BeautifulSoup    

```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://www.nytimes.com/section/us?action=click&pgtype=Homepage&region=TopBar&module=HPMiniNav&contentCollection=U.S.&WT.nav=page')
```

```python
show_btn = driver.find_element_by_class_name("buttons")
show_btn.click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser') 

title=[]

```

```python
for page in range(3,20) :
    titles = soup.find_all('h2',{'class':'headline'})
    for i in titles:
        title.append(i.text.strip())
    show_btn = driver.find_element_by_class_name("buttons")
    show_btn.click()
```

```python
print(title)
```

```python
summary= []
for i in soup.find_all('p',{'class':'summary'}):
    print(i.text)
    summary.append(i.text)
```

```python
driver.quit()
```
